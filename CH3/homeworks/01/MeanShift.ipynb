```python
import pandas as pd
import numpy as np
import csv
```

```python
checkins = pd.read_table(
    "checkins.dat",
    sep=r"\s*\|\s*",
    header=0,
    skiprows=[1],
    engine="python",
)
checkins.dropna(inplace=True)
```

```python
checkins = checkins.drop(checkins.columns[[0, 1, 2,5]], axis=1)
```

```python
from sklearn.cluster import MeanShift
checkins.info()
checkins_min = checkins[:1000]
```

```python
model = MeanShift(bandwidth=0.1)
model.fit(checkins_min)
```

```python
labels = model.labels_
cluster_centers = model.cluster_centers_
labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)

print("number of estimated clusters : %d" % n_clusters_)
```

```python
from collections import Counter
count = Counter(labels)
a = count.items()
```

```python
topcount = []
for n, item in enumerate(a):
    if item[1] > 15:
        topcount.append(item[0])
        
```

```python
#topcount номера необходимых массивов
#cluster_centers массив центров кластеров
points = [[33.751277, -118.188740], 
          [25.867736, -80.324116],
          [51.503016, -0.075479], 
          [52.378894, 4.885084],
          [39.366487, 117.036146 ],
          [-33.868457, 151.205134]]
allnearest = []
#numpy.linalg.norm

for n in zip(points):
    for z in zip(topcount):
        dist = np.linalg.norm(n-cluster_centers[z],ord=2)
        distitem = [n[0], [cluster_centers[z][0],cluster_centers[z][1]], dist]
        allnearest.append(distitem)
print distitem

        
```
