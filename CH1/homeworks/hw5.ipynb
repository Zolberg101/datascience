```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
import matplotlib.mlab as mlab
%matplotlib inline
```

# Нормальное распределение (50, 1)

```python
# выборка 
norm_rv = sts.norm(50, 1)
sample = norm_rv.rvs(1000)
```

```python
# теоретическая функция распределения и гистограмма:
x = np.linspace(45, 55,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')

plt.hist(sample, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

```python
#Выборки объема 5, 10, 50
samples5  = [norm_rv.rvs(5)  for _ in range(1000)]
samples10 = [norm_rv.rvs(10) for _ in range(1000)]
samples50 = [norm_rv.rvs(50) for _ in range(1000)]

sample5means  = [x.mean() for x in samples5]
sample10means = [x.mean() for x in samples10]
sample50means = [x.mean() for x in samples50]
```

```python
# мат. ожидание исходной случайной величины.
mean = 50.0

# По утверждению центральной предельной теоремы распределение выборочных средних достаточно хорошо описывается 
# нормальным распределением
```

```python
# вычислим стандартное отклонение для размера выборки в 5 и построим график

standard_deviation = math.sqrt(1 / 5.0)

x = np.arange(47, 53, 0.001)
plt.plot(x, sts.norm.pdf(x, mean, standard_deviation))

plt.hist(sample5means, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

```python
# вычислим стандартное отклонение для размера выборки в 10 и построим график

standard_deviation = math.sqrt(1 / 10.0)

x = np.arange(47, 53, 0.001)
plt.plot(x, sts.norm.pdf(x, mean, standard_deviation))

plt.hist(sample10means, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

```python
# вычислим стандартное отклонение для размера выборки в 50 и построим график

standard_deviation = math.sqrt(1 / 50.0)
x = np.arange(49, 51, 0.001)

plt.plot(x, sts.norm.pdf(x, mean, standard_deviation))

plt.hist(sample50means, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

### Выводы

Чем больше мы берем размер выборки тем плотнее значиния приближаються к имсходному мат. ожиданию выбраной случайной величины.
