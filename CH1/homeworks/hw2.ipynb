```python
%matplotlib inline
import numpy as np
from scipy import optimize
from scipy import linalg
from scipy import interpolate
from matplotlib import pylab as plt
```

```python
def graph(x):
    y = np.sin(x / 5.) * np.exp(x / 10.) + 5. * np.exp(-x / 2.)
    return y

x = np.arange(1,15,0.1)
y = graph(x)
plt.plot(x,y)
```

получение первого многочлена
    w1 + w2*x = f(x)
    w1 + w2*x = f(x)

```python
x1 = 1
x2 = 15
b1 = graph(x1)
b2 = graph(x2)

a = np.array([[1, 1], [1, 15]])
b = np.array([b1, b2])

z = linalg.solve(a, b)
print z
```

```python
plt.plot(x,y)
plt.plot([1,15], [graph(1),graph(15)])
plt.show()
```

```python
x1 = 1
x2 = 8
x3 = 15
b1 = graph(x1)
b2 = graph(x2)
b3 = graph(x3)

#w1 + w2x + w3x**2
a = np.array([[1, x1, x1**2],[1, x2, x2**2], [1, x3, x3**2]])
b = np.array([b1, b2, b3])
z = linalg.solve(a,b)
print z
```

```python
plt.plot(x,y)
plt.plot([1,8,15], [graph(1),graph(8),graph(15)])
plt.show()
```

```python
x1 = 1
x2 = 4
x3 = 10
x4 = 15
b1 = graph(x1)
b2 = graph(x2)
b3 = graph(x3)
b4 = graph(x4)

#w1 + w2x + w3x**2 + w4x**3
a = np.array([[1, x1, x1**2,x1**3],[1, x2, x2**2, x2**3], [1, x3, x3**2, x3**3],[1, x4, x4**2, x4**3]])
b = np.array([b1, b2, b3,b4])
z = linalg.solve(a,b)
print z
```
