```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from statsmodels.distributions.empirical_distribution import ECDF
%matplotlib inline
```

```python
# Задаем распределение
par = 0.5 # lambda
name = sts.expon(scale = 1 / par) # https://en.wikipedia.org/wiki/Exponential_distribution

# Генерируем выборку объема 1000
sample = name.rvs(1000)

# Гистограмма выборки
plt.hist(sample, bins = 30, normed = True, label = 'number of samples', alpha = 0.5)
plt.ylabel('$y(x)$')
plt.xlabel('$x$')

# Плотность распределения
x = np.linspace(min(sample), max(sample), 1000)
pdf = name.pdf(x)
plt.plot(x, pdf, label = 'theoretical pdf')
plt.legend(loc = 'best')
```

PDF of exponential distribution = lambda * exp(-lambda)
mu = 1 / lambda
sigma = 1 / (lambda ^ 2)


```python
# Function calculates means, draw hist, draw norm_rv with mu = 1 / lambda and sigma = 1 / lambda ^ 2 
def draw(n):
    samples = np.zeros([1000, n])
    S = np.zeros(1000)
    for i in range(1000):
        samples[i] = name.rvs(n)
        S[i] = np.mean(samples[i])
    plt.hist(S, bins = 30, normed = True, label = 'number of samples', alpha = 0.5)
    x = np.linspace(min(S), max(S), 100)
    mu = 1 / par
    sigma = (1 / (par ** 2 * n)) ** 0.5
    norm_rv = sts.norm(loc = mu, scale = sigma)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf)
    plt.xlabel('$x$')
    plt.ylabel('number of samples')
```

```python
draw(5)
```

```python
draw(10)
```

```python
draw(50)
```

Вывод: при увеличении числа n мы видим, что гистоаграмма выборки лучше приближается нормальным распределением, с соответсвующими параметрами, следоательно, ЦПТ работает.
